# Sorting lists

```python
x = [7,11,3,9,2]

print('sorted(x) =', sorted(x))
print('x =', x)
```

```python
x.sort()
print('x =', x)
```

```python
x = [7,11,3,9,2]
y = sorted(x, reverse=True)

print(y)
```

```python
x = [('a', 3), ('c', 1), ('b', 5)]
print(sorted(x))
```

Keyword `key` allows you to supply a (single arg) function, that compute some alternate value from an item. The items of list then sorted on basis of these alternate values.

Lambda notation allows us to create temporary functions (so that we don't have to define them, as previously shown).

```python
print(sorted(x, key=lambda i:i[1])) # sorting list of pairs (tuples) by second value
```

# Sorting dictionaries

Sorting a dictionary by key:

```python
counts = {'a': 3, 'c': 1, 'b': 5}

print('counts =', counts, end='\n\n')
print('sorted(counts) =', sorted(counts), end='\n\n')
print('sorted(counts, key=lambda v:counts[v]) =', sorted(counts, key=lambda v:counts[v]), end='\n\n')
print('sorted(counts, key=lambda v:counts[v], reverse=True) =', sorted(counts, key=lambda v:counts[v], reverse=True), end='\n\n')
```

Sorting a dictionary by value:

```python
densities = {'iron':7.8, 'gold':19.3, 'zinc':7.13, 'lead':11.4}

for metal in sorted(densities, reverse=True, key=lambda m:densities[m]):
    print('%8s = %5.1f' % (metal, densities[metal]))
```

```python
# alternative way to print, using f-strings

for metal in sorted(densities, reverse=True, key=lambda m:densities[m]):
    print(f'{metal: >8} = {densities[metal]: >5.1f}')
```
